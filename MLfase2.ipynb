```python
merge22 = pd.merge(df_produksi22, df_luaspanen22, on=["Kabupaten/Kota", "Tahun"], how="inner")
merge23 = pd.merge(df_produksi23, df_luaspanen23, on=["Kabupaten/Kota", "Tahun"], how="inner")
merge24 = pd.merge(df_produksi24, df_luaspanen24, on=["Kabupaten/Kota", "Tahun"], how="inner")
```

```python
merge22 = pd.merge(df_produksi22, df_luaspanen22, on=["Kabupaten/Kota", "Tahun"], how="inner")
merge23 = pd.merge(df_produksi23, df_luaspanen23, on=["Kabupaten/Kota", "Tahun"], how="inner")
merge24 = pd.merge(df_produksi24, df_luaspanen24, on=["Kabupaten/Kota", "Tahun"], how="inner")
```

```python
def bersihkan(df):
    df = df.replace("...", np.nan)
    df = df.fillna(0)
    return df
```

```python
merge22 = bersihkan(merge22)
merge23 = bersihkan(merge23)
merge24 = bersihkan(merge24)
```

```python
def clustering_tahunan(df, tahun, n_cluster=3):
    print(f"\n=== HASIL CLUSTERING TAHUN {tahun} ===")

    data = df.copy()
    kab = data["Kabupaten/Kota"]

    # drop kolom non-numerik
    data = data.drop(columns=["Kabupaten/Kota", "Tahun"])

    # pastikan semua kolom numeric
    data = data.apply(pd.to_numeric, errors="coerce").fillna(0)

    # normalisasi
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data)

    # K-Means
    kmeans = KMeans(n_clusters=n_cluster, random_state=42)
    labels = kmeans.fit_predict(data_scaled)

    hasil = pd.DataFrame({
        "Kabupaten/Kota": kab,
        f"Cluster_{tahun}": labels
    })

    print(hasil)
    return hasil
```

```python
cluster22 = clustering_tahunan(merge22, 2022)
cluster23 = clustering_tahunan(merge23, 2023)
cluster24 = clustering_tahunan(merge24, 2024)
```

```python
# ==== SATUKAN SEMUA TAHUN ====
dataset_final = pd.concat([merge22, merge23, merge24], ignore_index=True)
```

```python
# ==== SIMPAN SEBAGAI CSV FINAL ====
dataset_final.to_csv("dataset_final.csv", index=False)
```
